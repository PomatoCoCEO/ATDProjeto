```matlab
%%file representData.m
function representData(nameFile, nfigure)
    file = fopen(nameFile, 'r');
    vec = fscanf(file,'%f %f %f', [3 Inf]);
    vec = vec';

    %disp(vec(1,:))% imported
    %disp(vec(2,:))
    %disp(vec(end,:))
    %disp(size(vec,1))
    aid = size(vec,1);
    % the signals were captured at a frequency of 50 Hz -> T = 0.02s
    t = (0:0.02:((aid-1)*0.02))/60;
    figure(nfigure)
    subplot(311)
    plot(t, vec(:,1))
    title('x axis')
    xlabel('t[min]')
    ylabel('acc')
    subplot(312)
    plot(t, vec(:,2))
    title('y axis')
    xlabel('t[min]')
    ylabel('acc')
    subplot(313)
    plot(t, vec(:,3))
    title('z axis')
    xlabel('t[min]')
    ylabel('acc')
    end
```

```matlab
representData('acc_exp02_user01.txt',1)
representData('acc_exp01_user01.txt',2)
```
